```python
from flask import jsonify, request, Flask
import json
from flask_ngrok import run_with_ngrok

from summarizer import Summarizer
model = Summarizer()
app = Flask(__name__)
run_with_ngrok(app)


@app.route('/<string:q_id>', methods=['GET'])
def home(q_id):
    res = model.calculate_optimal_k(q_id, k_max=10)
    result = model(q_id, ratio=0.33)
    full = ''.join(result)
    L=full.split('.')
    if L=="":
      L=q_id
    return jsonify(list(L))


if __name__ == '__main__':
    app.run()
```
